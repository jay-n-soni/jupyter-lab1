```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([[0,20],[60,50]])
y=np.array([[0,250],[145,258]])
```

```python
plt.plot(x,y) # generate a line plot with x values on the horizontal axis and y values on the vertical axis,
plt.show() # show an output
```

```python
plt.plot(x,y,'*') # generate a marker point with x values on the horizontal axis and y values on the vertical axis,
plt.show() # show an output
```

```python
n1=np.array([10,20,40,55,25,654,546,785,128,8,864])
n2=np.array([60,65,818,856,254,185,28,2,721,687,18])
plt.plot(n1,n2) 
plt.show() # show an output
```

```python
plt.plot(n1,n2,'*')
plt.show() # show an output
```

```python
plt.plot(n2,marker='p',color='g',ms='10',mec='k',mfc='y',ls=":") # generate a line plot with marker, marker size, color, size, line style
plt.show() # show an output
```

```python
plt.plot(n1,marker='o',color='g',ls=':',lw='4') # generate a line plot with marker, color, size, line style, line width
plt.show()
```

```python
font={'color':'green','size':'20'} # initializing color and size of font
plt.title("Line Chart",fontdict=font,loc='right') # title of chart
plt.xlabel("Num 1") # x axis label
plt.ylabel("Num 2") # y axis label
plt.plot(n1,n2,marker='*') 
plt.grid(axis='y',color='r',ls='--',lw='.5') # display grid using axis, color, line style, and line width
```

```python
plt.subplot(2,2,1)
plt.title('p1')
plt.plot(n1,n2)
plt.subplot(2,2,2)
plt.title('p2')
plt.plot(n2,n1) # This will combine two graph and show in one super graph
plt.suptitle('Super') # Title of super graph 
```

```python
# x1=np.array([[12,564],[254,564]])
# y1=np.array([[458,564],[154,386]])
x1 = np.array([12, 564, 254, 564]) # made 1D
y1 = np.array([458, 564, 154, 386])
sizes=np.array([[10,50],[21,56]])
plt.scatter(x1,y1,color='red') # This will scatter a points in graphs 
plt.scatter(x,y,color='green',s=sizes,alpha=0.5) # This will scatter a points in graphs according to sizes 
plt.show()
```

```python
# x_bar=np.array([12,23,34,45,56])
plt.bar(x1,y1,color='m',width=100) #  creating bar graph 
plt.show()
```

```python
plt.barh(x1,y1,height=100) # creating bar graph horizontally
```

```python
plt.hist(x1) # creating histogram
```

```python
track1=np.array([25,54,36,48,54]) # array of percentage of each track
track_name=["data","CS","ML","AIML","WS"] # array of different track names
ex=[0,0.2,0,0,0] # Corrected list, now has 5 elements
col=["red","pink","m","g","b"] # colors given to different tracks

plt.pie(track1,labels=track_name,startangle=99,explode=ex,shadow=True,colors=col) # creating pie chart
plt.legend(title="Tracks") # creating legends of tracks in graph
plt.show()
```
